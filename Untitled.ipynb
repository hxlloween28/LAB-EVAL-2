```python
import pymongo
from bs4 import BeautifulSoup
import requests
```

```python
url='https://saras.cbse.gov.in/SARAS/AffiliatedList/ListOfSchdirReportNew?ID1=D'

data=requests.get(url)

parser= BeautifulSoup(data.content, 'html.parser')
parser
```

```python
t=parser.find_all('tr')
t
```

```python
column=[]

data={}
cnt=0
for i in t[1:]:
    i=str(i.text)
    st= i.split('\n')
    
    data[cnt]=st
    cnt=cnt+1
    
    
```

```python
data
```
